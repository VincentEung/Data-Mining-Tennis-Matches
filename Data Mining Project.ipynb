```python
import networkx as nx
import igraph as ig

# Sélectionner les joueurs spécifiques pour le clustering
# Sélectionner les trois joueurs
joueur1 = 'Rafael Nadal'
joueur2 = 'Roger Federer'
joueur3 = 'Novak Djokovic'

# Filtrer les données pour les trois joueurs sélectionnés
joueur1_data = df[df['winner_name'] == joueur1]
joueur2_data = df[df['winner_name'] == joueur2]
joueur3_data = df[df['winner_name'] == joueur3]

# Créer une liste de joueurs et de victoires
joueurs = [joueur1, joueur2, joueur3]

# Sélectionner les colonnes pertinentes pour le clustering
data_clustering = df[['winner_name', 'surface']]

# Créer le graphique à partir des données de clustering avec NetworkX
graph = nx.from_pandas_edgelist(data_clustering, 'winner_name', 'surface')

# Convertir le graphique NetworkX en un graphique igraph
g = ig.Graph.TupleList(graph.edges(), directed=False)

# Effectuer le clustering avec l'algorithme Walktrap
clusters = g.community_walktrap().as_clustering()

# Définir les paramètres de mise en page pour le dézoom
layout = nx.spring_layout(graph, k=0.1, scale=10)

# Visualiser le graphique avec les clusters et les nouveaux paramètres de mise en page
nx.draw_networkx(graph, pos=layout, node_color=clusters.membership, with_labels=True)
plt.show()

# Afficher les clusters
for i, cluster in enumerate(clusters):
    print(f"Cluster {i+1}: {cluster}")
```

```python
import networkx as nx
import igraph as ig
import matplotlib.pyplot as plt

# Sélectionner les joueurs spécifiques pour le clustering
joueur1 = 'Rafael Nadal'
joueur2 = 'Roger Federer'
joueur3 = 'Novak Djokovic'

# Filtrer les données pour les trois joueurs sélectionnés
joueur1_data = df[df['winner_name'] == joueur1]
joueur2_data = df[df['winner_name'] == joueur2]
joueur3_data = df[df['winner_name'] == joueur3]

# Créer une liste de joueurs et de victoires
joueurs = [joueur1, joueur2, joueur3]

# Sélectionner les colonnes pertinentes pour le clustering
data_clustering = df[(df['winner_name'].isin(joueurs)) | (df['loser_name'].isin(joueurs))]
data_clustering = data_clustering[['winner_name', 'surface']]

# Créer le graphique à partir des données de clustering avec NetworkX
graph = nx.from_pandas_edgelist(data_clustering,'winner_name', 'surface')

# Convertir le graphique NetworkX en un graphique igraph
g = ig.Graph.TupleList(graph.edges(), directed=False)

# Effectuer le clustering avec l'algorithme Walktrap
clusters = g.community_walktrap().as_clustering()

# Filtrer le graphique pour n'afficher que les joueurs spécifiques
graph_filtered = graph.subgraph(joueurs)

# Récupérer les couleurs des nœuds spécifiques
node_colors = [clusters.membership[joueurs.index(node)] for node in graph_filtered.nodes()]

# Visualiser le graphique avec les clusters et les nouveaux paramètres de mise en page
nx.draw_networkx(graph_filtered, pos=layout, node_color=node_colors, with_labels=True)
plt.show()

# Afficher les clusters
for i, cluster in enumerate(clusters):
    print(f"Cluster {i+1}: {cluster}")
```

Tout d'abord, le code utilisé demande les bibliothèques suivantes : NetworkX (“nx”) et iGraph (“ig”) pour créer et analyser nos données. Ainsi, les joueurs spécifiques pour le clustering sont définis dans les variables “joueur1”, “joueur2” et “joueur3”, qui sont respectivement "Rafael Nadal", "Roger Federer" et "Novak Djokovic".

On crée donc une liste “joueurs” est créée pour contenir les noms des joueurs spécifiques ("Rafael Nadal", "Roger Federer" et "Novak Djokovic"). (Cette liste sera beaucoup dans nos prochains exemples pour nous simplifier le travail)

Les colonnes pertinentes pour le clustering sont : "winner_name" (nom du joueur gagnant) et "surface" (type de surface du match). 

Or, un graphe NetworkX est créé à partir des données de clustering à l'aide de la fonction “nx.from_pandas_edgelist()”. Les nœuds du graphe représentent les joueurs et les arêtes du graphe représentent les relations entre les joueurs et les surfaces de leurs matchs.

Le graphe NetworkX va alors être converti en un graphe iGraph (“g”) à l'aide de la fonction “ig.Graph.TupleList()”.Puis nous procédons le clustering est effectué sur le graphe iGraph en utilisant l'algorithme Walktrap à l'aide de la fonction “g.community_walktrap().as_clustering()”.

Par la suite, le graphe NetworkX va être filtré pour n'afficher que les nœuds spécifiques pour les joueurs sélectionnés, en utilisant “graph_filtered = graph.subgraph(joueurs)”.

Les couleurs des nœuds spécifiques sont récupérées en utilisant la liste “node_colors”, qui est construite en parcourant chaque nœud dans le graphe filtré (“graph_filtered”) et en récupérant l'appartenance de chaque nœud au cluster à partir de “clusters.membership”. Enfin, le graphe filtré est visualisé avec les clusters et les nouvelles couleurs des nœuds en utilisant la fonction “nx.draw_networkx()”. Les noms des joueurs sont affichés sur les nœuds (“with_labels=True”). La boucle "for" permet aussi d'afficher les clusters.

L'utilisation de la variable “layout” permet l'utilisation des nœuds dans le graphique. Cependant, cette variable n'est pas définie dans le code fourni, ce qui peut causer une erreur.
```python
from sklearn.mixture import GaussianMixture

# Filtrer les données pour les joueurs Nadal et Federer
joueur1_data = df[df['winner_name'] == 'Rafael Nadal']
joueur2_data = df[df['winner_name'] == 'Roger Federer']

# Concaténer les DataFrames
joueurs_data = pd.concat([joueur1_data, joueur2_data], ignore_index=True)

# Sélectionner les colonnes pour le clustering
colonnes_clustering = ['winner_rank', 'winner_rank_points']  # Ajoutez d'autres colonnes pertinentes si nécessaire
data_clustering = joueurs_data[colonnes_clustering]

# Créer le modèle de mélange gaussien
gmm = GaussianMixture(n_components=2)  # Nous choisissons 2 composantes pour comparer les joueurs Nadal et Federer

# Entraîner le modèle de mélange gaussien
gmm.fit(data_clustering)

# Effectuer la prédiction des clusters
clustering_labels = gmm.predict(data_clustering)

# Associer chaque joueur à une couleur spécifique
couleurs_joueurs = {'Rafael Nadal': 'red', 'Roger Federer': 'blue'}

# Visualiser les clusters avec les noms des joueurs
plt.figure(figsize=(8, 6))
for i, joueur in enumerate(joueurs_data['winner_name']):
    plt.scatter(joueurs_data.loc[i, 'winner_rank'], joueurs_data.loc[i, 'winner_rank_points'],
                c=couleurs_joueurs[joueur], label=joueur)

plt.xlabel('Classement du joueur')
plt.ylabel('Points de classement du joueur')
plt.title('Comparaison des joueurs Nadal et Federer avec GMM Clustering')

# Ne montrer qu'une seule légende pour chaque couleur
handles, labels = plt.gca().get_legend_handles_labels()
by_label = dict(zip(labels, handles))
plt.legend(by_label.values(), by_label.keys(), bbox_to_anchor=(1.05, 1), loc='upper left', borderaxespad=0.)

plt.show()
```

```python
from sklearn.cluster import SpectralClustering

# Sélectionner les colonnes pertinentes pour le clustering (par exemple, les types de surface)
data_clustering = joueurs_data[['winner_name', 'surface']]

# Créer une table pivot pour compter le nombre de victoires de chaque joueur pour chaque type de surface
pivot_table = pd.pivot_table(data_clustering, index='winner_name', columns='surface', aggfunc=len, fill_value=0)

# Obtenir les types de terrains
surfaces = pivot_table.columns

# Obtenir les données pour le clustering
X = pivot_table.values

# Créer une matrice de similarité en utilisant la distance euclidienne entre les joueurs
from sklearn.metrics.pairwise import euclidean_distances
similarite_matrix = euclidean_distances(X)

# Effectuer le Spectral Clustering avec 2 clusters pour différencier les joueurs Nadal et Djokovic
spectral_clustering = SpectralClustering(n_clusters=2, affinity='precomputed', random_state=42)
clustering_labels = spectral_clustering.fit_predict(similarite_matrix)

# Associer chaque joueur à une couleur spécifique
couleurs_joueurs = {joueur1: 'red', joueur2: 'blue'}

# Visualiser les clusters avec les noms des joueurs
plt.figure(figsize=(12, 6))
for i, joueur in enumerate(pivot_table.index):
    x = np.arange(len(surfaces))
    y = X[i, :]
    plt.scatter(x, y, c=couleurs_joueurs[joueur], label=joueur)

plt.xticks(np.arange(len(surfaces)), surfaces)
plt.xlabel('Type de surface')
plt.title('Comparaison des joueurs Nadal et Djokovic avec Spectral Clustering')
plt.legend(loc='upper left')
plt.show()
```

Nous commencons par définir les noms des joueurs que nous souhaitons comparer en utilisant la méthode de Spectral Clustering. Dans cet exemple, nous choisissons 'Rafael Nadal' et 'Novak Djokovic'.

Puis nous sélectionnons les colonnes pertinentes pour effectuer le clustering. Dans cet exemple, nous choisissons les noms des joueurs ('winner_name') et les types de surfaces ('surface').

Par la suite nous créons une table pivot qui compte le nombre de victoires de chaque joueur pour chaque type de surface. Cela nous donne une matrice qui sera utilisée pour le clustering. On va alors ensuite mettre les différents types de terrains dans la variable surfaces. On obtient donc le cluster suivant dans lequel nous stockons les données dans la variable "X".

Nous calculons la distance euclidienne entre les joueurs à partir de la matrice X, ce qui nous donne une matrice de similarité.

Puis nous utilisons le Spectral Clustering avec 2 clusters sur la matrice de similarité similarite_matrix. Cela nous donne des étiquettes de clustering dans clustering_labels. Ensuite, nous créons un dictionnaire qui associe chaque joueur ('Rafael Nadal' et 'Novak Djokovic') à une couleur spécifique pour la visualisation.

Enfin, nous utilisons matplotlib pour créer le graphique. Nous utilisons scatter pour afficher les points pour chaque joueur en fonction du nombre de victoires sur chaque type de surface. Les couleurs des points sont déterminées par le dictionnaire couleurs_joueurs
. Nous ajoutons également les étiquettes des types de surfaces sur l'axe x et ajoutons un titre au graphique.
